```python
# Tushar.Tripathi or tushar.tripathi or Tushar.Tripathi@abc.com
def NormalizeUserName(userid):
  """
  Normalize the user id.
  If the user id is with domain name then it removes the domain name and return the user id
  """
  if "@" in  userid :
    return userid.split('@')[0].lower()
  else:
    return userid.lower()
```

```python
NormalizeUserName("tushar.tripathi@alphabeta.com")
```

```python
NormalizeUserName("tushar.tripathi123@alphabeta.com")
```

```python
NormalizeUserName("Tushar.Tripathi121@alphabeta.com")
```

```python
def CheckCreds(username,password,hasexpired):
  '''
  The first returned value => authentication status
  THe second return => account status
  '''
  if username == "admin":
    if password == "password":
      if hasexpired == True:
        return "authenticated","locked"
      return "authenticated","active"
  else:
    return "authentication failed",""

```

```python
response =  CheckCreds('admin','password',True)
```
